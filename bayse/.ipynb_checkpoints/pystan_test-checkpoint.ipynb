# 1.実際に乱数を生成し、その母数を推定する
pythonで既知の母数を持つ正規乱数を生成し、その結果のみを用いて母数を推定してみよう。<br>
stanがベイズ推論した母数は、真の母数と一致するだろうか？

```python
import pystan
import sklearn
import numpy as np
from numpy.random import *

y=normal(30,9,100) #期待値30、標準偏差9の正規分布に従う乱数を100個生成する
y
```

```python
# データを辞書型にする
dat = {'N': len(y), 'y': y}

# Stanコード。別ファイル（.stan）にdata{から書いてもよし。
#インデントを気にしない言語ではあるがちゃんとインデントしよう。
model = """
    data {
        int N;
        vector[N] y;
    }
    parameters { 
        real mu;
        real<lower=0> sigma;
    }
    model { 
        for (i in 1:N)
            y[i] ~ normal(mu,sigma);
    }
"""
```

```python
#モデルのコンパイル。C++コードに変換するにはpystan.StanModel関数。
stm = pystan.StanModel(model_code=model)
```

```python
n_itr = 5000 #MCMCの長さ
n_warmup = 1000 #バーンイン長さ
chains = 2 #2つの初期値で独立にMCMCサンプリングする

# MCMCサンプリングの実行。.samplingメソッド
fit = stm.sampling(data=dat, iter=n_itr, chains=chains, n_jobs=-1, warmup=n_warmup, algorithm="NUTS", verbose=False)
fit
```

結果表を見てみよう。特に注目するべきはmean(平均・つまり母数の推定値だ）とseである。<br>
一番右のRhatは収束しているかどうかの判定に使えるのだが、実際に乱数列をプロットしたものを見たほうが良い。

```python
import matplotlib.pyplot as plt
ext=fit.extract()
plt.plot(ext["mu"])
```

```python
import matplotlib.pyplot as plt
ext=fit.extract()
plt.plot(ext["sigma"])
```
